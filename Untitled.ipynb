```python
# Dependencies
import json
import requests
```

```python
# Specify the URL
url = "https://www.fema.gov/api/open/v1/HazardMitigationGrantProgramDisasterSummaries"

# Make request and store response
response = requests.get(url)

# Print status code
print(response.status_code)
```

```python
# JSON-ify response
response_json = response.json()
# json.dumps(response_json,indent = 2)
print(json.dumps(response_json, indent=2))
```
